```python
# Qiskit 2.x Example â€“ OpenQASM Demonstration

from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator
import qiskit.qasm3 as qasm3  # For OpenQASM 3 export
```

```python
# Create a simple 2-qubit quantum circuit
qc = QuantumCircuit(2)
qc.h(0)
qc.cx(0, 1)
qc.measure_all()
```

```python
print("Quantum Circuit:")
print(qc.draw())
```

```python
# Convert the circuit to OpenQASM 3 format
qasm_code = qasm3.dumps(qc)
print("\nGenerated OpenQASM 3 Code:\n")
print(qasm_code)
```

```python
# Optional: Simulate the circuit
simulator = AerSimulator()
compiled_circuit = transpile(qc, simulator)
result = simulator.run(compiled_circuit).result()
counts = result.get_counts()
print("\nSimulation Result:")
print(counts)
```

# **Programming Tasks**

***1. Modify the circuit to have 3 qubits and apply a different gate sequence.***

```python
from qiskit import QuantumCircuit
from qiskit.visualization import circuit_drawer

# Create a 3-qubit circuit
qc = QuantumCircuit(3)
qc.h(0)            # Hadamard on qubit 0
qc.cx(0, 1)        # CNOT from qubit 0 -> 1
qc.x(2)            # Pauli-X (NOT) on qubit 2
qc.cz(1, 2)        # Controlled-Z between qubits 1 and 2
qc.measure_all()

print("Modified 3-Qubit Quantum Circuit:")
print(qc.draw('text'))
```

***2. Generate and display the QASM code for the new circuit.***

```python
from qiskit import QuantumCircuit
import qiskit.qasm3 as qasm3

qc = QuantumCircuit(3)
qc.h(0)
qc.cx(0, 1)
qc.x(2)
qc.cz(1, 2)
qc.measure_all()

# Export to QASM 3
qasm_code = qasm3.dumps(qc)

print("Generated OpenQASM 3 Code:\n")
print(qasm_code)
```

***3. Identify how Hadamard and CNOT operations are represented in QASM syntax.***

```python
from qiskit import QuantumCircuit
import qiskit.qasm3 as qasm3

qc = QuantumCircuit(2)
qc.h(0)
qc.cx(0, 1)
qc.measure_all()

qasm_code = qasm3.dumps(qc)
print("QASM 3 Code:\n")
print(qasm_code)

print("\nğŸ”¹ Gate Representation in QASM 3:")
print("Hadamard gate (qc.h(0))  â†’  'h q[0];'")
print("CNOT gate (qc.cx(0,1))  â†’  'cx q[0], q[1];'")
```

***4. Add a barrier and observe how it appears in the QASM code.***

```python
from qiskit import QuantumCircuit
import qiskit.qasm3 as qasm3

qc = QuantumCircuit(3)
qc.h(0)
qc.cx(0, 1)
qc.barrier()      # <-- Barrier inserted here
qc.x(2)
qc.cz(1, 2)
qc.measure_all()

print("Quantum Circuit with Barrier:")
print(qc.draw('text'))

qasm_code = qasm3.dumps(qc)
print("\nGenerated QASM 3 Code:\n")
print(qasm_code)

print("\nğŸ”¹ Barrier in QASM 3 appears as:")
print("  'barrier q[0], q[1], q[2];'")
```

***5. Run the modified circuit and compare simulation results.***

```python
from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
import qiskit.qasm3 as qasm3
import matplotlib.pyplot as plt

# Create and modify the circuit
qc = QuantumCircuit(3)
qc.h(0)
qc.cx(0, 1)
qc.barrier()
qc.x(2)
qc.cz(1, 2)
qc.measure_all()

# Print QASM 3
print("Generated OpenQASM 3 Code:\n")
print(qasm3.dumps(qc))

# Simulate
sim = AerSimulator()
compiled = transpile(qc, sim)
result = sim.run(compiled, shots=2048).result()
counts = result.get_counts()

print("\nSimulation Results (Measurement Counts):")
print(counts)

plot_histogram(counts, title="QPE Task 5 â€“ Modified Circuit Results")
plt.show(block=True)
```
